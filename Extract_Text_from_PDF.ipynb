```
# run for the first time to install these packages if absent, then comment them out
'''!pip install pdf2image
!apt-get install tesseract-ocr
!apt-get install libtesseract-dev
!apt-get install pytesseract
!apt-get install poppler-utils'''

import pdf2image
# we are importing os,sys to manipulate paths and rename text files
import os, sys
# pytesseract is an OCR tool available in python 
import pytesseract


# PIL is a library which is used to work with images.
try:
    from PIL import Image
except ImportError:
    import Image


PATH = '/content/'
i=1

pdf_files =[]

# In this loop we are collecting all the pdf files from the given path  
for f in os.listdir(PATH):
    full_name = os.path.join(PATH, f)
    if os.path.isfile(full_name):
        name = os.path.basename(f)
        filename, ext = os.path.splitext(name)
        if ext == '.pdf':
            pdf_files.append(name)



# this function helps delete all the ppm images from our directory.
def delete_ppms():
    for file in os.listdir(PATH):
        if '.ppm' in file or '.DS_Store' in file:
            try:
                os.remove(PATH + file)
            except FileNotFoundError:
                pass

# to extract from multiple pdf files in given path
def pdf_extract_multiple(file, i):
    print('Extracting from :',file)
    delete_ppms() # delete all ppms before extraction to avoid mixing ppms of different file
    images = pdf2image.convert_from_path(pdf_path=PATH + file,output_folder=PATH)
    j=0
    for file in sorted(os.listdir(PATH)):
        if '.ppm' in file and 'image' not in file:
            os.rename(PATH + file,PATH + 'image' + str(i) + '-' + str(j) + '.ppm') 
            j += 1
    j=0
    f = open(PATH + 'result{}.text'.format(i),'w')
    files = [f for f in os.listdir(PATH) if '.ppm' in f]

    for file in sorted(files, key=lambda x: int(x[x.index('-') +1: x.index('.')])):
        temp = pytesseract.image_to_string(Image.open(PATH+file))
        f.write(temp)
    f.close()

# to extract from a single file
def pdf_extract(file):
    print('Extracting from ',file)
    delete_ppms() # delete all ppms before extraction to avoid mixing ppms of different file

    images = pdf2image.convert_from_path(pdf_path =PATH+file, output_folder=PATH)
    
    page_no=0 # 
    for file in sorted(os.listdir(PATH)):
        if '.ppm' in file and 'image' not in file:
            # we are renaming pdf images because after converting to ppms ,names are  like this 
            # /content/032d3f8e-ab32-4734-8fd9-1fc2ecb0583b-01.ppm ->  /content/image-0.ppm
            os.rename(PATH+file, PATH+'image'+'-'+str(page_no)+'.ppm')
            page_no+=1

    page_no=0  #reset j
    
    #opening a new text file to store the extracted text 
    f = open(PATH + 'extracted.text','w')
    
    # files is a list to get all the ppm image names from the given path
    files = [f for f in os.listdir(PATH) if '.ppm' in f]
    #print(files)

    '''
    We are applying a lambda function to sort our files 
    as our files have name like  => image-x.ppm , we are sorting them using 
    numbers given to the ppm images. This is done to avoid messing up the sequence
    of the pdf pages.
    '''
    for file in sorted(files, key=lambda x: int(x[x.index('-')+1 : x.index('.')])):
        temp = pytesseract.image_to_string(Image.open(PATH+file))
        #Here temp is a page from pdf and we are converting that image to text 
        #print('-----------------------------------------\n',temp)
        f.write(temp)
    
    delete_ppms() # deleting all the ppm images which we created to extract from each page.  
    f.close    

print(pdf_files)

pdf_extract('Essay Writing Guide.pdf')

'''for i in range(len(pdf_files)):
    pdf_file = pdf_files[i]
    pdf_extract_multiple(pdf_file,i)
    print('done')
'''
```
